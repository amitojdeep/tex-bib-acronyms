```python
# source file
with open('main.bib', 'r') as file:
    data = file.read()
```

```python
new_data = ""
i = 0
while i < len(data): 
    
    #stop a bit before the end, copy last 10 as is
    if(i == len(data)-20):
        new_data+=data[i:i+20]
        #print("last few chars" + data[i:i+20])
        break
        
    # find all combinations of author part
    if(data[i:i+8] == "author={" or data[i:i+10] == "author = {" or data[i:i+9] == "author= {" 
       or data[i:i+9] == "author ={"):
        
        #continue till the end bracket
        while(data[i] != '}'):
            
                #put the spaces, "and", lastnames 
                while(data[i] != ',' and data[i] != '}'):
                    new_data = new_data+data[i]
                    #print("Inside author: " + data[i])
                    i+=1
                
                # for lastnames with chars like {\'e}
                    if(data[i] == '}' and data[i-4] == "{" and data[i-2] == "\'"):
                        #new_data = new_data+data[i]
                        print("Found and included in lastname : " + data[i-4:i+1] )
                        i+=1
                
                    elif(data[i] == '}'):
                        #add the }
                        new_data = new_data + data[i]
                        #print("Closing braces:" + data[i])
                        i+=1
                        break
                        
                #append comma
                new_data = new_data+data[i]
                #print("Comma: " + data[i])
                i+=1
                
                #add space after comma if there
                
                if(data[i] == ' '):
                    new_data = new_data+data[i]
                    #print("Space post comma: " + data[i])
                    i+=1
            
                # append the first letter of first name
                new_data = new_data+data[i]
                #print("First letter:" + data[i])
                i+=1
                
                #check if name was already shortened and had a middle name
                # cuases error if first two letters of first name are caps
                while(data[i] >= 'A' and data[i] <= 'Z'):
                    new_data = new_data+data[i]
                    #print("Already shortened middle name:" + data[i])
                    i+=1
                
                # skip rest of the characters of first name
                while(data[i] != ' ' and data[i] != '}' ):
                    #print("skipped: " + data[i])
                    i+=1
                    
                    # for firstnames with chars like {\'e}
                    if(data[i] == '}' and data[i-4] == "{" and data[i-2] == "\'"):
                        #new_data = new_data+data[i]
                        #print("Found and skipped in first name: " + data[i-4:i+1] )
                        i+=1
                    elif(data[i] == '}'):
                        break
                            
            
                
                # append the space after first name if it exists and continue
                if(data[i] == ' '):
                    new_data = new_data+data[i]
                    #print("space after first name" + data[i])
                    i+=1
                
                #check for middle name post the space after the first name
                if(data[i:i+3] != 'and' and data[i] != '}'):
                    #middle name exists, enter first char
                    new_data = new_data+data[i]
                    #print("First letter middle name:" + data[i])
                    i+=1                
                    #check if middle name was pre shortened and had multiple caps => Mike, John AX
                    # can cause issues
                    while(data[i] >= 'A' and data[i] <= 'Z'):
                        new_data = new_data+data[i]
                        #print("Capital in shortened middle name:" + data[i])
                        i+=1
                
                    # skip rest of the characters of middle name
                    while(data[i] != ' ' and data[i] != '}'):
                        #print("skipped: " + data[i])
                        i+=1
        
    else:
        new_data = new_data+data[i]
        #print("Outside author:" + data[i])
        i+=1
        
print(new_data)
```

```python
#destination
with open("new.bib", "w") as text_file:
    text_file.write(new_data)
```
